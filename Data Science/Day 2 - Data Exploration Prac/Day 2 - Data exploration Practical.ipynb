**Next the data has to be normalised**

But we can't just jump in and normalise, we need to think about it first.

**Make a copy of the dataframe.**

```python
data = df.copy()
data.head()
```

The unique id would not typically be used in a prediction and the date can not be normalised.

**Remove the id colunm using 'drop'.**

```python
data = data.drop(['id'], axis=1)
data.head()
```

We can also not use the date in the current format since it is not a float or integer. There are several ways to address this.

You can seperate the year, month and day into seperate columns.

Or if you think that that level of detail is not needed you can only use the year.

**Convert the date column into seperate year, month and day colunms**

Step 1. Convert the 'date' col to a datetime format.

```python
data['date'] = pd.to_datetime(data['date'])
data.head()
```

Step 2. Extract the year from the date column (which should now be in a datetime format). And put it in a new column named 'year'.

```python
data['year'] = data['date'].dt.year
data.head()
```

Step 3. Repeat this for the month and day.

Finally visualise the data.

```python
data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day
data.head()
```

If you look at the far right then you should be able to see the year, month and day columns.

The date column is still in the dataframe, so let's remove it.

**Remove the 'date' column**



```python
data = data.drop(['date'], axis=1)
data.head()
```

**Normalise the data**

NOTE: there are MANY ways to do this. The method discussed in the presentation is a manual way to normalise specific columns but there are other automated methodologies that you can try out.

If you want to try one of the functions mentioned during the presentation, Google "sklearn.preprocessing.normalize()" or "sklearn.preprocessing.MinMaxScaler()" to find the literature and examples.

```python
for column in list(data.columns):
  data[column] = (data[column]-data[column].min())/(data[column].max()-data[column].min())
data.head()
```
