## **VISUALISATION**

**Get a list of the available seaborn datasets.**

```python
import seaborn as sns
sns.get_dataset_names()
```

**Load the "penguins" seaborn dataset and have a quick look at it (use 'display' or 'print').**

```python
data = sns.load_dataset("penguins")
print(data)
```

There are quite a few NaN values in this dataset, but since we are only focussig on visualisation for this session, let's ignore them....

But now we have a sence of the different variables that we can plot.

# ***LINE PLOTS***

**Use matplotlib to plot the 'bill length' and the 'flipper length' on the same line graph.**

```python
import matplotlib.pyplot as plt

plt.plot(data['bill_length_mm'], label='Bill(length)')
plt.plot(data['flipper_length_mm'], label='Flipper(length)')
plt.grid()
plt.xlabel('Sample')
plt.ylabel('Length(mm)')
plt.title('Plotting Bill Lenth & Flipper Length for Penguins')
plt.legend()
```

**Now plot the 'body mass' and the 'flipper length' on the same line graph.**



```python
plt.plot(data['body_mass_g'], label="Body Mass")
plt.plot(data['flipper_length_mm'], label = "Flipper Length")
plt.grid()
plt.xlabel('Sample')
plt.ylabel('Units(g and mm)')
plt.title('Body Mass & Flipper Length for Penguins')
plt.legend()
```

Well, that doesn't look good. And the units of each variable is different, which can easily confuse a reader.

**Plot the 'body mass' and the 'flipper length' again in the same graph but on different y-axis.**

```python
fig, ax = plt.subplots()


ax.set_xlabel('Sample')
ax.plot(data['flipper_length_mm'], color = 'indigo', label = "Flipper Length")
ax.set_ylabel('Flipper Length', color = "indigo")

ax2 = ax.twinx()
ax2.plot(data['body_mass_g'], color = 'red', label="Body Mass")
ax2.set_ylabel("Body Mass", color='red')
ax2.set_title('Plotting Body Mass & Flipper Lenght with different Units')
ax2.grid()
```

# ***SCATTER PLOTS***

**Plot the 'body mass' versus the 'flipper length' on a scatter plot.**

```python
plt.scatter(data['body_mass_g'], data['flipper_length_mm'], color = 'magenta')
plt.xlabel('Mass(grams)')
plt.ylabel('Length(mm)')
plt.title('Body Mass VS Flipper Length')
```

The relationship between these 2 variables can now easily be deduced.

# ***HISTOGRAMS***

**Create a histogram of the 'body mass' with 80 bins.**

```python
plt.hist(data['body_mass_g'], bins=80)
plt.ylabel('Number of sampes')
plt.xlabel('Mass')
plt.title('Mass Distribution')
plt.grid()
```

This looks kinda noisy. It looks like there
seems to be between 7 and 9 'clusters'.
Let's play with the number of bins.

**Plot a new histogram with xx bins.**

```python
plt.hist(data['body_mass_g'], bins=30)
plt.ylabel('Number of sampes')
plt.xlabel('Mass')
plt.title('Mass Distribution')
plt.grid()
```

That is better. The most penguins seems to have a body mass between about 3200 and 4000 grams.

**Recreate the same histogram with 8 bins using seaborn (with the kernel density).**

```python
sns.set_style('darkgrid')
sns.histplot(data['body_mass_g'], bins=8, kde=True)
plt.ylabel('Number of sampes')
plt.xlabel('Mass')
plt.title('Mass Distribution')
plt.grid()
```

The peak seems to be around 3600 grams.

# ***HEAT MAPS***

**Create a heatmap looking at 'cut' and 'color'.** Hint: use groupby and have a look to make sure that the data looks correct.

```python
df_m = df.groupby(["cut","color"]).size().unstack(level=0)
sns.heatmap(df_m, annot=True, fmt='.4g')
```
